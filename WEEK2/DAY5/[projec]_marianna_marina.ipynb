```python
import pandas as pd
```

```python
data_attacks = pd.read_csv('attacks.csv', encoding='latin-1')
```

```python
#No total temos 25723 ataques na tabela. As duas ultimas colunas, que sÃ£o nomeadas como Unamed: 22 e Unamed: 23 
#possuem mais de 99% dos seus valores nulos, por isso decidimos que faz sentido para a limpeza dos dados apagar 
#essas duas colunas


data_attacks.drop('Unnamed: 22', axis=1, inplace = True)
data_attacks.drop('Unnamed: 23', axis=1, inplace = True)
```

```python
data_attacks.dropna(axis=0, how='all', subset=['href'], inplace = True)

## ApÃ³s ter aberto o arquivo csv, verificamos que a partir de determinado ponto, as linhas nao continham 
#informacoes uteis para o data frame, por estarem faltando ou sendo nulas. Portanto, decicimos excluir estas 
#linhas.
```

```python
data_attacks.rename(columns={'Sex	':'Sex'}, inplace = True)
data_attacks.rename(columns={'Species	':'Species'}, inplace = True)
```

```python
data_attacks['Name'] = data_attacks['Name'].str.replace('female','Unknown')
data_attacks['Name'] = data_attacks['Name'].str.replace('male','Unknown')

# Em alguns casos em vez de colocar o nome da pessoa foi posto o gÃªnero, por isso decidimos alinhar o dado de forma
# que a informaÃ§Ã£o esteja coerente com a coluna, assim todas as vezes em que o nome esteja como male ou female, o nome serÃ¡
# modificado para Unknown. 
```

```python

data_attacks = data_attacks.fillna('Unknown')
##Optamos pela opÃ§Ã£o de preencher os missing values 
## por outro valor das colunas 
##no caso o pela string Unknown. Pois, caso
##os valores dessa coluna fossem apagados 
##nÃ³s tambÃ©m perderiamos as informaÃ§oes das 
##outras colunas, o que nÃ£o seria 
##algo indesejÃ¡vel, pois muitas informaÃ§Ãµes 
##seriam perdidas
```

```python
dif_list=[]
dif_list2=[]
dif_list3=[]
dif_list4=[]

for i in range(len(data_attacks)):
    if data_attacks['href formula'][i]!=data_attacks['href'][i]:
        dif_list.append(i)
    if data_attacks['Case Number.1'][i]!=data_attacks['Case Number.2'][i]:
        dif_list2.append(i)
    if data_attacks['Case Number'][i]!=data_attacks['Case Number.1'][i]:
        dif_list3.append(i)
    if data_attacks['Case Number'][i]!=data_attacks['Case Number.2'][i]:
        dif_list4.append(i)
        
print(len(dif_list))
print(len(dif_list2))
print(len(dif_list3))
print(len(dif_list4))

data_attacks = data_attacks.drop('Case Number.1', axis=1)
data_attacks = data_attacks.drop('Case Number.2', axis=1)
data_attacks = data_attacks.drop('href formula', axis=1)


# Aparentemente, as colunas continham valores iguais, apÃ³s anÃ¡lise por meio de um for loop e verificaÃ§Ã£o manual, 
# foi possÃ­vel identificar que, de fato, as diferenÃ§as eram muito poucas e decidimos manter apenas uma coluna com
# a informaÃ§Ã£o.
```

```python
morning = ['5', '6', '7', '8', '9', '10', '11']
afternoon = ['12', '13', '14', '15', '16', '17', '18']
night = ['19', '20', '21', '22', '23', '24', '0', '1', '2', '3', '4']

for i in range(len(data_attacks)):
    for time in morning:
        if  time in data_attacks['Time'][i]:
            data_attacks['Time'][i] = 'Morning'
            
    for time in afternoon:
        if  time in data_attacks['Time'][i]:
            data_attacks['Time'][i] = 'Afternoon'
            
    for time in night:
        if  time in data_attacks['Time'][i]:
            data_attacks['Time'][i] = 'Night'
            
# Para organizar a coluna dos horarios optamos em tentar organizar os ataques pelos tres turnos existentes: 
# manha, tarde e noite. Por isso todas as linhas que continham informaÃ§Ã£o numÃ©rica alteramos para a forma de 
# âmorningâ, âafternoonâ, or ânightâ
```

```python
data_attacks['Time'] = data_attacks['Time'].replace(['Late afternoon','Midday','Early afternoon','Evening','P.M.','Dusk','After%'],'Afternoon')


data_attacks['Time'] = data_attacks['Time'].replace(['1Morning','0Morning','Early morning','A.M.'],'Morning')

```

```python
data_attacks = data_attacks.drop('pdf', axis=1)

## Decidimos deletar a coluna com os dados do pdf, pois essa coluna consiste em informaÃ§Ãµes que 
## se repetem em outras colunas como o nÃºmero do caso, nome, localizaÃ§Ã£o, etc
```

```python
data_attacks['Sex'].replace(['M ', 'lli', 'N', '.'], 'Unkwown')
data_attacks['Fatal (Y/N)'].replace(['M', 'UNKNOWN', '2017'], 'Unkwown')

#FunÃ§Ãµes feitas para substituir os valores considerados errados nas listas e substituidos por desconhecido
```

```python
data_attacks = data_attacks.drop('Year', axis=1)

## as informaÃ§Ãµes contidas na coluna Date possuem os valores dos anos, tornando-se assim redundante trabalhar
## tambÃ©m com a coluna 'year'. Decidimos deletar a coluna com os dados do yeat, pois essa coluna consiste em informaÃ§Ãµes 
## menos organizadas e completas quando comparadas com a coluna 'date'.
```

```python
for i in range(len(data_attacks)):
    if i < 10:
        data_attacks['Case Number'][i] = f'CN_0000{i}'
    if i < 100:
        data_attacks['Case Number'][i] = f'CN_000{i}'
    if i < 1000:
        data_attacks['Case Number'][i] = f'CN_00{i}'
    else:
        data_attacks['Case Number'][i] = f'CN_0{i}'
    
# Para tornar um index confiÃ¡vel padronizamos a coluna do numero do caso da segunte forma 
# CN_OO(nÃºmero da linha)
```
