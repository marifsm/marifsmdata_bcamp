```python
import requests
import json
from bs4 import BeautifulSoup
import feedparser

#Nosso primeiro passo foi importar todas as bibliotecas necessárias para o nosso programa
```

O primeiro site que escolhemos para fazer o webscraping foi o Adorocinema, nosso objetivo é selecionar categorias de título, direçao, elenco, ratings, etc...


```python
adorocinema1 = 'http://www.adorocinema.com/filmes/numero-cinemas/'
page = requests.get(adorocinema1)
```

Agora usamos a função page.status_code para verificar se o link esta respondendo e que será factível para o uso do web scraping:

```python

page.status_code
```

```python
soup = BeautifulSoup(page.content, 'html.parser')
```

O primeiro parametro escolhido para fazer o web scraping será o título dos filmes, assim  analisamos na pagina para encontrar a classe. A segunda categoria escolhida foi a data de estreia do filme, então novamente nós procuramos a classe das datas. A terceira categoria utilizada para fazer o web scraping foi o diretor do filme, para isso mais uma vez identificamos a classe e limpamos o ruido remanescente do texto na célula seguinte

```python

title = soup.find_all('a',class_="meta-title-link")
date = soup.find_all('span',class_="date")
director = soup.find_all('a',class_="blue-link")


```

Depois nos limpamos as listas através da função split()

```python
title_t = [title[i].get_text().split(">") for i in range(len(title))]
title_t = [word for i in title for word in i]
date_t = [date[i].get_text().split(">") for i in range(len(date))]
date_t = [word for i in date for word in i]
director_t = [director[i].get_text().split(">") for i in range(len(director))]
director_t = [word for i in director for word in i]
```

O proximo passo é criar um loop para fazer o mesmo procedimento nas próximas páginas:

```python
title2 = []
date2 = []
director2 = []

for i in range(2,5):
    adorocinema2 = f'http://www.adorocinema.com/filmes/numero-cinemas/?page={i}'
    page2 = requests.get(adorocinema2)
    soup2 = BeautifulSoup(page2.content, 'html.parser')
    
    title2.append(soup2.find_all('a',class_="meta-title-link"))
    t2 = [word for i in title2 for word in i]
    t2_t = [t2[i].get_text().split(">") for i in range(len(t2))]
    t2_t = [word for i in t2_t for word in i]
    date2.append(soup2.find_all('span',class_="date"))
    d2 = [word for i in date2 for word in i]
    d2_t = [d2[i].get_text().split(">") for i in range(len(d2))]
    d2_t = [word for i in d2_t for word in i]
    director2.append(soup2.find_all('a',class_="blue-link"))
    dr2 = [word for i in director2 for word in i]
    dr2_t = [dr2[i].get_text().split(">") for i in range(len(dr2))]
    dr2_t = [word for i in dr2_t for word in i]
    
```

```python
for i in range(len(t2_t)):
    title_t.append(t2_t[i])
    date_t.append(d2_t[i])
    director_t.append(dr2_t[i])
```

Devido à impossibilidades técnicas em relação à formatação das informações do site, não conseguimos extrair mais informações de colunas do site, por isso decidimos criar o dataframe e fazer webscraping em outro site:

```python
import pandas as pd

adorocinema_df = {
        'Título': (title_t),
        'Data de lançamento': (date_t),
        'Direção': (director_t),
}
    
    
df_info = pd.DataFrame(adorocinema_df, columns = ['Título', 'Data de lançamento','Direção'])
```

```python
df_info
```

```python
df_info.to_csv('df_adorocinema.csv')
```

Como nosso segundo objetivo iremos fazer webscraping no site de notícas, o site escolhido foi o UOL. Dessa vez optamos fazer a captação de dados através do RSS, que usa formatos de feed de internet padrão para publicar informações atualizadas com freqüência, o que combina bastante com o nosso caso de utilizar um blog de notícias.

O documento RSS  inclui Texto resumido e metadados, como data de publicação, link de acesso ao conteúdo e nome do autor. O leitor de RSS verifica os feeds regularmente para obter novas informações e pode baixá-lo automaticamente, se essa função estiver ativada. O leitor também fornece uma interface de usuário, que o caso dos navegadores modernos. 


```python
uol = feedparser.parse('http://rss.home.uol.com.br/index.xml')

#para isso primeiro acessamos o link do blog de noticias UOL em RSS
```

Em seguida geramos o dataframe com todas as informações que o documento RSS nos forneceu e para acessar as informações, geramos uma visualização com a função .head().

Muitas das informações fornecidas não foram usadas, o critério utilizado foi focado mais nas informações que seriam interessantes serem usadas e que estariam disponiveis visualmente no site ao ser acessado do que nas informaçoes que acreditamos que deveriam ser excliudas. Por isso, apos analisar o documento, decidios trabalahar com os dados do título, resumo da notícia, data da publicação e créditos da noticia.

```python
df_uol = pd.DataFrame(uol.entries)
df_uol.head(2)
```

```python
df_uol['Tamanho do titulo'] = df_uol['title'].apply(len)
df_final = df_uol[['title','Tamanho do titulo']].sort_values('Tamanho do titulo', ascending=False)
```

```python
df_final['Resumo'] = df_uol['summary']
df_final['Data da publicação'] = df_uol['published']
df_final ['Creditos da noticia'] = df_uol ['credit']
df_final.columns = ['Titulo', 'Tamanho do titulo', 'Resumo', 'Data da publicação', 'Creditos da noticia']
```

```python
df_final.to_csv('df_uol.csv')
```

Para o nosso terceiro site fazendo web scrape decidimos fazer em um site de previsão de tempo:

```python
climatempo = 'https://www.climatempo.com.br/previsao-do-tempo/15-dias/cidade/558/saopaulo-sp'
page = requests.get(climatempo)
soup = BeautifulSoup(page.content, 'html.parser')
```

Nos decidimos retirar as informações de temperatura máxima, temperatura mínima, dia/local...

```python

t_max = soup.find_all('span',class_="max-temp")
t_min = soup.find_all('span',class_="min-temp")
dia = soup.find_all('p',class_="-gray _float-l _max-w-75-pct")
```

Porém, assim como nos outros casos, as informações obtidas com o soup.find_all não estão 100% "limpas", elas possuem ruidos das informaçoes das classes com as informações contidas no texto do HTML. Por isso, o próximo passo para gerar o dataframe é utilizar string functions para organizar as informaçoes e gerar um dataframe sem ruidos.

```python
t_maxt = [t_max[i].get_text().split(">") for i in range(len(t_max))]
tmx_t = [word for i in t_maxt for word in i]

t_mint = [t_min[i].get_text().split(">") for i in range(len(t_min))]
tmn_t = [word for i in t_mint for word in i]

dia_t = [dia[i].get_text().replace("\n"," ").split(">") for i in range(len(dia))]
d_t = [word for i in dia_t for word in i]
```

Como a informação do "dia_t" contém três informações inclusas (dia da semana, data e local) decidimos usar a separação de vírgula para separar as informações em variáveis distintas para que as mesmas sejam colunas diferentes no dataframe.

```python
dia_t = [dia[i].get_text().replace("\n"," ").split(">") for i in range(len(dia))]
d_t = [word for i in dia_t for word in i]

week = []
day = []
place = []

for i in range(len(d_t)): 
    week.append(d_t[i].split(",")[0])
    day.append(d_t[i].split(",")[1])
    place.append(d_t[i].split(",")[2])
```

```python


climatempo_dic = {
        'Semana': (week),
        'Dia': (day),
        'Local': (place),
        'Temperatura máxima': (tmx_t[:15]),
        'Temperatura mínima': (tmn_t[:15]),
}
    
    
df_temp = pd.DataFrame(climatempo_dic, columns = ['Temperatura máxima','Temperatura mínima', 'Semana', 'Dia', 'Local'])
```

```python
df_temp
```

```python
df_temp.to_csv('df_climatempo.csv')
```

Nosso próximo web scraping também é mais um site de noticias, dessa vez usaremos o G1 stire de notícias e seguiremos o mesmo workflow de todos os outros sites:

```python
g1 = 'https://g1.globo.com/'
page = requests.get(g1)
soup = BeautifulSoup(page.content, 'html.parser')
```

```python
title = soup.find_all('a',class_="feed-post-link gui-color-primary gui-color-hover")
summary = soup.find_all('a',class_="gui-color-primary gui-color-hover feed-post-body-title bstn-relatedtext")
tipo = soup.find_all('span',class_="feed-post-header-chapeu")
```

```python
title_t = [title[i].get_text().split(">") for i in range(len(title))]
title_final = [word for i in title_t for word in i]

summary_t = [summary[i].get_text().split(">") for i in range(len(summary))]
summary_final = [word for i in summary_t for word in i]

tipo_t = [tipo[i].get_text().split(">") for i in range(len(tipo))]
tipo_final = [word for i in tipo_t for word in i]
```

```python
g1_dic = {
        'Titulo': (title_final)[:6],
        'Resumo': (summary_final),
        'Tipo de noticia': (tipo_final)[:6],
        
}
    
    
df_g1 = pd.DataFrame(g1_dic, columns = ['Titulo','Resumo', 'Tipo de noticia'])
```

```python
df_g1['Tamanho do titulo'] = df_g1['Titulo'].apply(len)
df_g1[['Titulo','Tamanho do titulo']].sort_values('Tamanho do titulo', ascending=False)
```
